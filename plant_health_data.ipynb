```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('plant_health_data.csv')
df.head(5)
```

```python
df.drop(["Timestamp","Plant_ID"], axis=1)
```

```python
X = df.drop(["Timestamp","Plant_ID","Plant_Health_Status"], axis=1)
y = df["Plant_Health_Status"]
```

```python
Y=df["Plant_Health_Status"]
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
y.shape
```

```python
num_cols = X.select_dtypes(include=np.number).columns.tolist()

num_pipline = Pipeline(steps=[
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipline, num_cols)
])
```

```python
preprocessor
```

```python
prepocesser_obj = preprocessor.fit_transform(X)
```

```python
prepocesser_obj.shape
```

```python
tress_level_mapping = {'High Stress': 0, 'Moderate Stress': 1, 'Healthy': 2}


y = list(map(lambda x: tress_level_mapping.get(x, x), Y))
```

```python
y= np.array(y)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(prepocesser_obj, y, test_size=0.2)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
param_ra ={
        "class_weight":["balanced"],
        'n_estimators': [20, 50, 30],
        'max_depth': [10, 8, 5],
        'min_samples_split': [2, 5, 10],
        }

params_de = {
            "class_weight":["balanced"],
            "criterion":['gini',"entropy","log_loss"],
            "splitter":['best','random'],
            "max_depth":[3,4,5,6],
            "min_samples_split":[2,3,4,5],
            "min_samples_leaf":[1,2,3],
            "max_features":["auto","sqrt","log2"]
            }

params_lo = {
            "class_weight":["balanced"],
            'penalty': ['l1', 'l2'],
            'C': [0.001, 0.01, 0.1, 1, 10, 100],
            'solver': ['liblinear', 'saga']
            }
```

```python
clf_rf = RandomForestClassifier()
clf_de = DecisionTreeClassifier()
clf_lo = LogisticRegression()

grid = GridSearchCV(clf_rf, param_ra, cv=5, scoring='accuracy', verbose=1 , n_jobs=-1)
grid.fit(X_train, y_train)
best_params_rf = grid.best_params_
best_rf = RandomForestClassifier(**best_params_rf)
best_rf.fit(X_train, y_train)
y_pred_rf = best_rf.predict(X_test)
y_pred_t = best_rf.predict(X_train)
accuracy_rf_t = accuracy_score(y_train, y_pred_t)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy test: {accuracy_rf}")
print(f"Random Forest Accuracy train: {accuracy_rf_t}")
```
